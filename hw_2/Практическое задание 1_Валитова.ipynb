```python
class Chocolate:
  def __init__(self, prace:float, additive:str =None): # Инициация объекта,задание типа данных и необъязательности 2 переменной
    if prace <0:    #Проверка цены на положительность
      print ("ValueError")
    self.prace = prace
    self.additive = additive
  def show_my_chocolate(self): # определение и вывод типа батончика
    if self.additive == None:
        return print ("Обычный шоколадный батончик")
    else:
        return print ("Шоколадный батончик, добавка:", self.additive )
  def __str__(self): # вывод стоимости батончика
    return f'{"Шоколадный батончик"} {self.prace} {"рублей"}'
```

```python
coco = Chocolate(3.12 ) # Пример без добавки шоколада
```

```python
coco = Chocolate(1, "арахис") # Пример положительная цена с добавкой
```

```python
coco = Chocolate(-1, "арахис") # Пример отрицательная цена
```

```python
coco.show_my_chocolate()
```
